# 01 — Data Exploration (Phase 0 sanity check)

Quick confirmation that the built `ProblemInstance` looks right: demand distribution, top SKUs, demand per region, demand over the 12 periods, and a scatter of facility + region coordinates. **Sanity check, not analysis.**

```python
import os, sys
sys.path.insert(0, os.path.join(os.path.dirname(os.getcwd()), 'src'))
sys.path.insert(0, os.path.join(os.getcwd(), 'src'))
import pandas as pd
import matplotlib.pyplot as plt
import config
from instance import ProblemInstance

inst = ProblemInstance.load(config.INSTANCE_PATH)
inst.summary()
```

## Demand distribution (per cell)

```python
d = inst.demand
ax = d['demand'].plot(kind='hist', bins=50, title='Demand per (SKU,region,period) cell')
ax.set_xlabel('units'); plt.show()
d['demand'].describe()
```

## Top SKUs by total demand

```python
top = d.groupby('sku')['demand'].sum().sort_values(ascending=False)
top.plot(kind='bar', figsize=(12,4), title='Total demand per SKU'); plt.show()
top.head(10)
```

## Demand per region

```python
inst.demand_by_region().sort_values(ascending=False).plot(
    kind='bar', figsize=(10,4), title='Total demand per region'); plt.show()
```

## Demand over the 12 periods

```python
inst.demand_by_period().plot(
    marker='o', figsize=(10,4), title='Total demand per period'); plt.show()
```

## Facility + region coordinates

```python
fig, ax = plt.subplots(figsize=(7,7))
for r in inst.regions:
    lat, lon = inst.coords[r]; ax.scatter(lon, lat, c='tab:blue'); ax.annotate(r, (lon, lat))
for p in inst.plants:
    lat, lon = inst.coords[p]; ax.scatter(lon, lat, c='tab:red', marker='s', s=80); ax.annotate(p, (lon, lat))
for dc in inst.dcs:
    lat, lon = inst.coords[dc]; ax.scatter(lon, lat, c='tab:green', marker='^', s=80); ax.annotate(dc, (lon, lat))
ax.set_xlabel('lon'); ax.set_ylabel('lat'); ax.set_title('Regions (blue), Plants (red), DCs (green)')
plt.show()
```
